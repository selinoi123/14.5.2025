```python
df1 = pd.read_csv("C:/Users/Selin/Downloads/Items (1).csv") 
df2 = pd.read_csv("C:/Users/Selin/Downloads/customers (1).csv") 
df3 = pd.read_csv("C:/Users/Selin/Downloads/order_item (1).csv") 
df4 = pd.read_csv("C:/Users/Selin/Downloads/orders (1).csv") 
```

```python
df_items = df1.copy()
df_customers = df2.copy()
df_order_item = df3.copy()
df_orders = df4.copy()
```

```python
#Question num 2
#Article num 3
df_mergeds = pd.merge(df_order_item, df_items, left_on='item_id', right_on='id', how='left')
df_mergeds['total_price'] = df_mergeds['item_price'] * df_mergeds['quantity']

df_order_item['total_price'] = df_mergeds['total_price']
df_order_item
```

```python
#graphs
#Article num 1
gender_counts = df_customers['gender'].value_counts()
x = gender_counts.index
y = gender_counts.values

plt.figure(figsize=(15, 5))
plt.bar(x, y, color='skyblue')

plt.title('Customer Segmentation by Gender')
plt.xlabel('Gender')
plt.ylabel('Number of Customers')
plt.grid(axis='y', linestyle='--')
plt.tight_layout()
plt.show()
```

```python
#graphs
#Article num 3
plt.subplots(figsize=(7, 5))
x_age = df_customers['age'].unique()

plt.title('Customer segmentation by age')
plt.xlabel('age')
plt.ylabel('the number of customers')

plt.hist(x_age, bins = 20, color = 'red')
plt.tight_layout()
plt.grid(linestyle='--')
plt.show()
```
